```python
import numpy as np
import matplotlib.pyplot as plt

v1 = np.array([1, 2])  # Vektor [2, 3]
v2 = np.array([3, 5])  # Vektor [4, 6]
v3 = np.array([4, 4])

# Buat panah
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector v2')
plt.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color="yellow", label="vektor3 bro")

# Tambahkan panah ke plot
plt.xlim(0, 10)  # Tentukan batas x
plt.ylim(0, 10)  # Tentukan batas y
plt.grid(True)
plt.axhline(0, color='gray',linewidth=0.5)
plt.axvline(0, color='gray',linewidth=0.5)

plt.legend()  # Label vektor
plt.title('Visualization of Vectors')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
