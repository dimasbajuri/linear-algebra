# Dot Product dan Cross Product

## 1. Dot Product
Dot product (perkalian titik) dari dua vektor $\mathbf{u}$ dan $\mathbf{v}$ didefinisikan sebagai:
$$\mathbf{u} \cdot \mathbf{v} = u_1 v_1 + u_2 v_2 + u_3 v_3$$
Dimana:
- $u_1, u_2, u_3$ adalah komponen dari vektor $\mathbf{u}$,
- $v_1, v_2, v_3$ adalah komponen dari vektor $\mathbf{v}$.

**Interpretasi Geometris**:
- Dot product mengukur sejauh mana dua vektor searah.
- Hasil dot product adalah skalar yang mengindikasikan sudut antara dua vektor dan panjang vektor.
- Formula sudut antara dua vektor:
$$\cos \theta = \frac{\mathbf{u} \cdot \mathbf{v}}{\|\mathbf{u}\| \|\mathbf{v}\|}$$

### Contoh:
Misalkan:
$$\mathbf{u} = [1, 2, 3], \quad \mathbf{v} = [4, 5, 6]$$
Dot product:
$$\mathbf{u} \cdot \mathbf{v} = 1 \times 4 + 2 \times 5 + 3 \times 6 = 32$$

---

## 2. Cross Product
Cross product (perkalian silang) dari dua vektor $\mathbf{u}$ dan $\mathbf{v}$ didefinisikan sebagai:
$$\mathbf{u} \times \mathbf{v} = \begin{vmatrix} \mathbf{i} & \mathbf{j} & \mathbf{k} \\ u_1 & u_2 & u_3 \\ v_1 & v_2 & v_3 \end{vmatrix}$$
Dimana:
- $\mathbf{i}, \mathbf{j}, \mathbf{k}$ adalah unit vektor basis dari 3D,
- $u_1, u_2, u_3$ adalah komponen dari vektor $\mathbf{u}$,
- $v_1, v_2, v_3$ adalah komponen dari vektor $\mathbf{v}$.

**Interpretasi Geometris**:
- Cross product menghasilkan vektor normal yang tegak lurus terhadap kedua vektor $\mathbf{u}$ dan $\mathbf{v}$.
- Panjang cross product sebanding dengan luas segitiga yang dibentuk oleh vektor $\mathbf{u}$ dan $\mathbf{v}$.
- Hasil cross product menunjukkan arah dengan menggunakan aturan tangan kanan.

### Contoh:
Misalkan:
$$\mathbf{u} = [1, 2, 3], \quad \mathbf{v} = [4, 5, 6]$$
Cross product:
$$\mathbf{u} \times \mathbf{v} = \begin{vmatrix} \mathbf{i} & \mathbf{j} & \mathbf{k} \\ 1 & 2 & 3 \\ 4 & 5 & 6 \end{vmatrix} = [-3, 6, -3]$$

---

## 3. Perbedaan Intuisi
- **Dot Product** digunakan untuk:
  - Menentukan sudut antara vektor.
  - Mengukur panjang vektor.
  - Memeriksa kemiringan dan kesamaan arah.
  
- **Cross Product** digunakan untuk:
  - Mendapatkan vektor normal dari dua vektor.
  - Mengukur luas segitiga atau parallelogram yang dibentuk oleh dua vektor.
  - Arah vektor hasil cross product mengikuti aturan tangan kanan.
